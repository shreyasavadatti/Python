```python
#matplotlib plotting
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np
x=np.arange(0,10)
y=np.arange(11,21)
#scatter
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Graph.png')
```

```python
y=x*x
#plt plot 
#plt.plot(x,y,'b')
#plt.plot(x,y,'b--')
plt.plot(x,y,'ro--')
```

```python
y=x*x
#plt plot 
#plt.plot(x,y,'b')
#plt.plot(x,y,'b--')
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
#Creating subplot
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np
y=x*x
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
#computing x and y coordinates for points on a sine curve 
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
#subplot
#compute x and y coordinates for points on sine and cosine curve
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
#setting up subplot grid height as 2 and width 1
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')
```

```python
#bar plot 
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
#Histograms 
a=np.array([22,87,5,43,56,73,55,54,11,20,5,5,79,31,27])
plt.hist(a,bins=20)
plt.title('Histogram')
```

```python
#box plot
data=[np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot 
plt.boxplot(data,vert=True,patch_artist=True);
#patch artist=colour 
#vert=vertical
```
