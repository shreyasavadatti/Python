```python
import csv

# Data to be written to the CSV file
data = [
    ["username", "password"],
    ["user1", "pass123"],
    ["user2", "user@1234"],
    ["user3", "pass_1234"],
]

csv_file_path = "user_passwords.csv"


with open(csv_file_path, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV file created and data written successfully.")

def read_data():
    file=list(csv.reader(open("user_passwords.csv")))
    temp=[]
    for i in file:
        temp.append(i)
    return temp
def create_userid(temp):
    while True:
        userID=input("Enter a new user ID: ")
        userID.lower()
        inlist=False
        row=0
        for y in temp:
            if userID in temp[row][0]:
                print(userID, "has already been allocated")
                inlist=True
            row=row+1
        if inlist==False:
            break
    return userID
```

```python
def create_password():
    while True:
        special = ["!", "@", "#", "$", "%", "^", "&", "*", "(", ")", "?", "{", "}", "<", ">", "/", "\""]
        a = False
        b = False
        c = False
        d = False
        score = 0
        password = input("Enter your password: ")
        password2 = input("Confirm your password: ")  
        if len(password) >= 8:
            score += 1
        for x in password:
            if x.isupper():
                a = True
            if x.islower():
                b = True
            if x in special:
                c = True
            if x.isdigit():
                d = True

        if a:
            score += 1
        if b:
            score += 1
        if c:
            score += 1
        if d:
            score += 1

        if score == 1 or score == 2:
            print("This is a weak password, try again")
        elif score == 3 or score == 4:
            print("This password could be improved")
            again = input("Do you want to try again? (y/n)").lower()
            if again == 'n':
                break
        if password != password2:
            print("Passwords do not match. File not saved")
            main()  
        else:
            return password
```

```python
def find_user(temp):
    while True:
        search = input("Enter the User ID you are searching for: ")
        search = search.lower() 
        inlist = False
        row = 0
        userID = None 

        for y in temp:
            if search in y[0]:  
                inlist = True
                userID = y[0]  
                break  

            row = row + 1

        if inlist:
            break  
        else:
            print(search, "is not in the list")

    return userID
```

```python
def change_password(userID, temp):
    if userID != "":
        password = create_password()

        for user_data in temp:
            if user_data[0] == userID:
                user_data[1] = password

        with open("user_passwords.csv", "w") as file:
            for user_data in temp:
                new_record = f"{user_data[0]}, {user_data[1]}\n"
                file.write(new_record)
                
    else:
        print("User ID cannot be empty")
```

```python
def main():
    temp=read_data()
    while True:
        print()
        print("1) Create a new USER ID")
        print("2) Change a password")
        print("3) Display all USER IDs")
        print("4) Quit")
        print()
        sel=int(input("Enter your choice:"))
        if sel==1:
            userID=create_userid(temp)
            password=create_password()
            file=open("user_passwords", "a")
            newrec=userID+","+password+"\n"
            file.write(str(newrec))
            file.close()
        elif sel==2:
            userID=find_user(temp)
            change_password(userID,temp)
        elif sel==3:
            displayall_userID()
        elif sel==4:
            break
        else:
            print("Incorrect choice")
main()
```
