```python
#SEABORN
#anazlyzing multiple features
#for continous we have 
#distplot
#joinplot=bivariate analysis 
#pairplot
```

```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
#here tip is my dependent variable and others are my independent variable 
```

```python
#HEATMAP=helps to find correlation with each and every feature. Important in Feature Selection
#correlation is only found if values are int or float not categorical 
df.corr()
```

```python
#if total bill is increasing tip is increasing 
```

```python
sns.heatmap(df.corr())
```

```python
#JOINTPLOT
#allows to study relation between 2 numeric variables. Central chat display their correlation. 
```

```python
sns.jointplot(x='tip', y='total_bill', data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#PAIRPLOT
#MORE THAN TWO VARIABLES 
#ALSO KNOWN AS SCATTERPLOT,shows all variables paid with the other variables 
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
#DISTPLOT HELPS US TO CHECK DISTRIBUTION OF THE COLUMNS FEATURE 
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#cATEGORICAL PLOTS for categorical features
#boxplot,violinplot,countplot,barplot
```

```python
#COUNT PLOT shows the count of observations in each categorical bin using bars
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#BARPLOT
sns.barplot(x='total_bill', y='sex',data=df)
```

```python
#BOXPLOT(box and whisker plot)
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,orient='v')#considering all features
```

```python
sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
```

```python
#VIOLIN PLOT
#helps us to see both the dist of data in terms of kernel density estimation and box plot
```

```python
sns.violinplot(x="total_bill",y="day",data=df,palette='rainbow')
```
